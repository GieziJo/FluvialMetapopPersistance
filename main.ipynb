Load libraries

```python
import Source.AnalyseMultipleOCNs as multi_ocn
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
```

Generate results

```python
mo = multi_ocn.AnalyseMultipleOCNs()
```

```python
nSteps = 9
mo.launchAnalysis(nSteps)
```

Treat results for figures

```python
areas = np.zeros((len(mo.las), nSteps, len(mo.las[0].OCN.A.flatten())))
mean_areas = np.zeros((len(mo.las), nSteps))
fractionConnected = np.zeros((len(mo.las), nSteps))
sumLength = np.zeros((len(mo.las), nSteps))
LMs = np.zeros((len(mo.las), nSteps))
LM_vects = np.zeros((len(mo.las), nSteps, len(mo.las[0].OCN.A.flatten())))
varyingAreas = {}
for la_i in range(len(mo.las)):
    la = mo.las[la_i]
    ocn = la.OCN
    baseArea = np.sum(la.areaThr[0] / la.OCN.cellsize**2)
    for n_i in range(nSteps):
        varyingAreas[la_i, n_i] = la.areaThr[n_i]
        mean_areas[la_i, n_i] = np.mean(la.areaThr[n_i])
        sumLength[la_i, n_i] = np.sum(la.areaThr[n_i] > 0)
        fractionConnected[la_i, n_i] = sumLength[la_i, n_i] / (la.OCN.dimX * la.OCN.dimY)
        LMs[la_i, n_i] = la.LM[n_i]
        
        areas[la_i, n_i,:] = la.areaThr[n_i].flatten()
        l = la.LM_vect[n_i].flatten()
        l = l**2
        l = l / sum(l)
        LM_vects[la_i, n_i,:] = l*la.LM[n_i]
```
